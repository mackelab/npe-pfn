```python
from functools import partial

import matplotlib.pyplot as plt
import seaborn as sns

from tabpfn_sbi.utils import (
    multi_plot,
    plot_metric_by_num_simulations,
    plot_calibration_curves,
    query,
    use_style,
)
from tabpfn_sbi.utils.plot_utils import DEFAULT_COLORS
```

```python
name = "../../main_results/sbibm"
query(name, metric="c2st").head(10)
```

```python
tasks = [
    "gaussian_mixture",
    "two_moons",
    "slcp",
    "gaussian_linear",
    "sir",
    "bernoulli_glm",
    "lotka_volterra",
]
methods = ["npe", "nle", "tabpfn", "filtered_tabpfn"]

with use_style("pyloric"):
    fig, axes = multi_plot(
        "../../main_results/sbibm",
        "task",
        "metric",
        partial(plot_metric_by_num_simulations, hue="method"),
        verbose=False,
        metric="c2st",
        task=tasks,
        method=methods,
        figsize_per_col=1.0,
        figsize_per_row=1.4,
        color_map=DEFAULT_COLORS,
        hue="method",
        y_lims=(0.5, 1.03),
        legend_bbox_to_anchor=(0.58, 0.02),
    )

    axes[0, 0].set_ylabel("C2ST", x=-0.18)

    for ax in axes[0, 0:]:
        ax.spines["left"].set_bounds(0.5, 1)

    for ax in axes[0, 1:]:
        ax.set_yticklabels([])
        ax.set_xticklabels([])

    axes[0, 0].set_xlabel("Num. sim.")
    for ax in axes[0, 1:]:
        ax.set_xlabel("")

    fig.savefig("fig2_amortized.svg", bbox_inches="tight", transparent=True)
```

```python
tasks = [
    "gaussian_mixture",
    "two_moons",
    "slcp",
    "gaussian_linear",
    "sir",
    "bernoulli_glm",
    "lotka_volterra",
]
methods = [
    "npe",
    "nle",
    "nre",
    # "tabpfn",
    "filtered_tabpfn",
    "npe_ensemble",
]

with use_style("pyloric"):
    fig, axes = multi_plot(
        "../../main_results/sbibm",
        "task",
        "metric",
        partial(plot_metric_by_num_simulations, hue="method", alpha=0.5),
        verbose=False,
        metric="c2st",
        task=tasks,
        method=methods,
        figsize_per_col=1.05,
        figsize_per_row=1.2,
        color_map=DEFAULT_COLORS,
        hue="method",
        fig_legend=False,
        y_lims=(0.5, 1.03),
    )
    for ax in axes[0, 1:]:
        ax.set_yticklabels([])

    for ax in axes[0, 0:]:
        ax.set_xticklabels([])
        ax.set_xlabel("")

    # turn off the legend
    axes[0, 0].legend().remove()

    for ax in axes[0, 0:]:
        ax.spines["left"].set_bounds(0.5, 1)

    axes[0, 0].set_ylabel("C2ST", x=-0.4)
    fig.savefig("fig2_amortized_extended.svg", bbox_inches="tight", transparent=True)
```

runtimes

```python
eval_runtimes_df = query(name, method=["npe", "nle", "tabpfn", "filtered_tabpfn"])
eval_runtimes = eval_runtimes_df[["method", "num_simulations", "time_eval", "task"]]
eval_runtimes = eval_runtimes[
    (eval_runtimes["task"] == "two_moons")
    | (eval_runtimes["task"] == "gaussian_mixture")
    | (eval_runtimes["task"] == "slcp")
    | (eval_runtimes["task"] == "gaussian_linear")
    | (eval_runtimes["task"] == "bernoulli_glm")
    # | (eval_runtimes["task"] == "sir")
    # | (eval_runtimes["task"] == "lotka_volterra")
]

train_runtimes_df = query(name, method=["npe", "nle"])
runtimes_train = train_runtimes_df[["method", "num_simulations", "time_train", "task"]]
runtimes_train = runtimes_train[
    (runtimes_train["task"] == "two_moons")
    | (runtimes_train["task"] == "gaussian_mixture")
    | (runtimes_train["task"] == "slcp")
    | (runtimes_train["task"] == "gaussian_linear")
    | (runtimes_train["task"] == "bernoulli_glm")
    # | (runtimes_train["task"] == "sir")
    # | (runtimes_train["task"] == "lotka_volterra")
]

with use_style("pyloric"):
    fig, axes = plt.subplots(1, 2, figsize=(2.2, 0.9))
    _ = sns.swarmplot(
        data=eval_runtimes,
        x="num_simulations",
        y="time_eval",
        hue="method",
        ax=axes[0],
        size=1.3,
        legend=False,
        palette=DEFAULT_COLORS,
    )
    axes[0].set_yscale("log")
    axes[0].set_xticklabels([r"$10^2$", r"$10^3$", r"$10^4$", r"$10^5$"])
    axes[0].set_xlabel("Num. sim.")
    axes[0].set_ylabel("Time [s]")
    axes[0].set_title("Sampling time", y=1.05)
    axes[0].set_ylim(1e-2, 1e5)
    axes[0].set_yticks([0.1, 10.0, 1000, 100_000])

    _ = sns.swarmplot(
        data=runtimes_train,
        x="num_simulations",
        y="time_train",
        hue="method",
        ax=axes[1],
        size=1.3,
        legend=False,
        palette=DEFAULT_COLORS,
    )
    axes[1].set_yscale("log")
    axes[1].set_xticklabels([r"$10^2$", r"$10^3$", r"$10^4$", r"$10^5$"])
    axes[1].set_xlabel("")
    axes[1].set_ylabel("")
    axes[1].set_title("Training time", y=1.05)
    axes[1].set_ylim(1e-2, 1e5)
    axes[1].set_yticks([0.1, 10.0, 1000, 100_000])
    axes[1].set_xticklabels([])
    axes[1].set_yticklabels([])

    fig.savefig("fig2_runtime.svg", bbox_inches="tight", transparent=True)
```
