```python
import svgutils.transform as sg
from IPython.display import SVG


# create new SVG figure
fig = sg.SVGFigure(width="14cm", height="5.8cm")
# Subpanels
fig1 = sg.fromfile("fig2_amortized.svg")
plot1 = fig1.getroot()

plot2 = sg.fromfile("fig2_runtime.svg")
plot2 = plot2.getroot()

plot3 = sg.fromfile("fig2_sequential.svg")
plot3 = plot3.getroot()


plot1.moveto(13, 5, scale=1)
plot2.moveto(-5, 113, scale=1)
plot3.moveto(167, 113, scale=1)


# Text
txt1 = sg.TextElement(5, 15, "a", size=15, weight="bold", font="Arial")
txt2 = sg.TextElement(5, 120, "b", size=15, weight="bold", font="Arial")
txt3 = sg.TextElement(170, 120, "c", size=15, weight="bold", font="Arial")
fig.append([plot1, plot2, plot3])
fig.append([txt1, txt2, txt3])

fig.save("fig2.svg")
svg = SVG("fig2.svg")
svg
```
