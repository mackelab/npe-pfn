```python
from functools import partial

from tabpfn_sbi.utils import multi_plot, plot_metric_by_num_simulations, use_style
from tabpfn_sbi.utils.plot_utils import DEFAULT_COLORS
```

```python
tasks = [
    "gaussian_mixture",
    "two_moons",
    "slcp",
    "gaussian_linear",
    "sir",
    "bernoulli_glm",
    "lotka_volterra",
]
methods = ["npe", "npe_sweeper", "filtered_tabpfn", ""]

with use_style("pyloric"):
    fig, axes = multi_plot(
        "../../results/sbibm_sweeper",
        "task",
        "metric",
        partial(plot_metric_by_num_simulations, hue="method"),
        verbose=False,
        metric="c2st",
        task=tasks,
        method=methods,
        figsize_per_col=1.0,
        figsize_per_row=1.4,
        color_map=DEFAULT_COLORS,
        hue="method",
        y_lims=(0.5, 1.03),
        legend_bbox_to_anchor=(0.58, 0.02),
    )

    axes[0, 0].set_ylabel("C2ST", x=-0.18)

    for ax in axes[0, 0:]:
        ax.spines["left"].set_bounds(0.5, 1)

    for ax in axes[0, 1:]:
        ax.set_yticklabels([])
        ax.set_xticklabels([])

    axes[0, 0].set_xlabel("Num. sim.")
    for ax in axes[0, 1:]:
        ax.set_xlabel("")

    fig.savefig("npe_with_sweep.svg", bbox_inches="tight", transparent=True)
```
