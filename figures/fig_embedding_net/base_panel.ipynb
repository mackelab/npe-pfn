```python
from tabpfn_sbi.tasks.spatial_sir import SpatialSIRTask
import matplotlib.pyplot as plt
import torch

torch.manual_seed(42 * 8)

task = SpatialSIRTask(N_grid=22, T=6)


prior = task.get_prior_dist()
simulator = task.get_simulator()


theta_o = prior.sample()
x_o = simulator(theta_o)

with use_style("pyloric"):
    fig = plt.figure(figsize=(0.8, 0.8))
    ax = plt.gca()
    ax.imshow(x_o.reshape(22, 22), cmap="Reds")
    ax.axis("off")
    fig.savefig("spatial_sir_observation.svg", bbox_inches="tight", dpi=300)
```

```python
samples_npe = model_npe1k.sample((5000,), x=x_o.reshape(1, 1, 22, 22))
```

```python
samples = model.sample((5000,), x=x_o.flatten())
```

```python
x_pred = simulator(samples[:100])
```

```python
import seaborn as sns

prior_samples = prior.sample((10000,))
with use_style("pyloric"):
    fig = plt.figure(figsize=(1, 1))
    ax = plt.gca()
    sns.kdeplot(
        x=prior_samples[:, 0],
        y=prior_samples[:, 1],
        color="black",
        levels=10,
        alpha=0.1,
        ax=ax,
    )
    sns.kdeplot(
        x=samples[:, 0],
        y=samples[:, 1],
        color="navy",
        levels=500,
        shade=True,
        shade_lowest=False,
        alpha=0.5,
        ax=ax,
    )
    ax.scatter(theta_o[0], theta_o[1], color="red", alpha=0.5)
    ax.set_xlabel(r"infection rate")
    ax.set_ylabel(r"recovery rate")
    plt.xlim(0, 4)
    plt.ylim(0, 4)
    fig.savefig("spatial_sir_posterior.svg", bbox_inches="tight")
```

```python
from sbi.analysis.plot import marginal_plot, pairplot
from tabpfn_sbi.utils.plot_utils import DEFAULT_COLORS

task = get_task("lotka_volterra_long", flatten=False)
x_o = task.get_observation(7)
with use_style("pyloric"):
    fig, ax = plt.subplots(2, figsize=(2.5, 1.3))

    # plt.plot(xs_mean_npe.T)
    ax[0].plot(
        xs_q05_npe[0].T, "-", color=DEFAULT_COLORS["npe"], alpha=1.0, linewidth=0.1
    )
    ax[0].plot(
        xs_q95_npe[0].T, "-", color=DEFAULT_COLORS["npe"], alpha=1.0, linewidth=0.1
    )

    # plt.plot(xs_mean_tabpfn.T)
    ax[0].plot(
        xs_q05_tabpfn[0],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn"],
        alpha=1.0,
        linewidth=0.1,
    )
    ax[0].plot(
        xs_q95_tabpfn[0],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn"],
        alpha=1.0,
        linewidth=0.1,
    )

    # plt.plot(xs_mean_tabpfn_infomax.T)
    ax[0].plot(
        xs_q05_tabpfn_infomax[0],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn_infomax"],
        alpha=1.0,
        linewidth=0.1,
    )
    ax[0].plot(
        xs_q95_tabpfn_infomax[0],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn_infomax"],
        alpha=1.0,
        linewidth=0.1,
    )

    ax[0].plot(x_o[0, 0].T, ".", color="black", markersize=1.0)

    ax[0].set_xticks([])

    ax[1].plot(
        xs_q05_npe[1].T, "-", color=DEFAULT_COLORS["npe"], alpha=1.0, linewidth=0.1
    )
    ax[1].plot(
        xs_q95_npe[1].T, "-", color=DEFAULT_COLORS["npe"], alpha=1.0, linewidth=0.1
    )

    ax[1].plot(
        xs_q05_tabpfn[1],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn"],
        alpha=1.0,
        linewidth=0.1,
    )
    ax[1].plot(
        xs_q95_tabpfn[1],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn"],
        alpha=1.0,
        linewidth=0.1,
    )

    ax[1].plot(
        xs_q05_tabpfn_infomax[1],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn_infomax"],
        alpha=1.0,
        linewidth=0.1,
    )
    ax[1].plot(
        xs_q95_tabpfn_infomax[1],
        "-",
        color=DEFAULT_COLORS["filtered_tabpfn_infomax"],
        alpha=1.0,
        linewidth=0.1,
    )

    ax[1].plot(x_o[0, 1].T, ".", color="black", markersize=1)

    ax[1].set_xlabel("time")
    ax[1].set_ylabel("prey")
    ax[0].set_ylabel("predator")

    ax[0].set_yticks([1.0, 1.5])
    ax[1].set_yticks([0.5, 1.5])

    fig.savefig("lv_long_marginal_plot.svg", bbox_inches="tight")
```

```python
import numpy as np
from scipy.stats import gaussian_kde

# Create x values for smooth line plot
x = np.linspace(-1, 1, 100)

with use_style("pyloric"):
    fig, ax = plt.subplots(
        2, 2, figsize=(2, 1.0), gridspec_kw={"wspace": 0.4, "hspace": 1.1}
    )
    ax = ax.flatten()

    # Plot smoothed density curves
    kde_ref = gaussian_kde(reference_posterior_samples[:, 0])
    kde_tabpfn = gaussian_kde(samples_tabpfn[:, 0])
    kde_npe = gaussian_kde(samples_npe[:, 0])
    kde_tabpfn_infomax = gaussian_kde(samples_tabpfn_infomax[:, 0])

    ax[0].plot(x, kde_ref(x), color="black")
    ax[0].plot(x, kde_tabpfn(x), color=DEFAULT_COLORS["filtered_tabpfn"])
    ax[0].plot(x, kde_npe(x), color=DEFAULT_COLORS["npe"])
    ax[0].plot(
        x, kde_tabpfn_infomax(x), color=DEFAULT_COLORS["filtered_tabpfn_infomax"]
    )
    ax[0].set_xticklabels([])

    kde_ref = gaussian_kde(reference_posterior_samples[:, 1])
    kde_tabpfn = gaussian_kde(samples_tabpfn[:, 1])
    kde_npe = gaussian_kde(samples_npe[:, 1])
    kde_tabpfn_infomax = gaussian_kde(samples_tabpfn_infomax[:, 1])

    ax[1].plot(x, kde_ref(x), color="black")
    ax[1].plot(x, kde_tabpfn(x), color=DEFAULT_COLORS["filtered_tabpfn"])
    ax[1].plot(x, kde_npe(x), color=DEFAULT_COLORS["npe"])
    ax[1].plot(
        x, kde_tabpfn_infomax(x), color=DEFAULT_COLORS["filtered_tabpfn_infomax"]
    )
    ax[1].set_xticklabels([])

    kde_ref = gaussian_kde(reference_posterior_samples[:, 2])
    kde_tabpfn = gaussian_kde(samples_tabpfn[:, 2])
    kde_npe = gaussian_kde(samples_npe[:, 2])
    kde_tabpfn_infomax = gaussian_kde(samples_tabpfn_infomax[:, 2])

    ax[2].plot(x, kde_ref(x), color="black")
    ax[2].plot(x, kde_tabpfn(x), color=DEFAULT_COLORS["filtered_tabpfn"])
    ax[2].plot(x, kde_npe(x), color=DEFAULT_COLORS["npe"])
    ax[2].plot(
        x, kde_tabpfn_infomax(x), color=DEFAULT_COLORS["filtered_tabpfn_infomax"]
    )

    kde_ref = gaussian_kde(reference_posterior_samples[:, 3])
    kde_tabpfn = gaussian_kde(samples_tabpfn[:, 3])
    kde_npe = gaussian_kde(samples_npe[:, 3])
    kde_tabpfn_infomax = gaussian_kde(samples_tabpfn_infomax[:, 3])

    ax[3].plot(x, kde_ref(x), color="black")
    ax[3].plot(x, kde_tabpfn(x), color=DEFAULT_COLORS["filtered_tabpfn"])
    ax[3].plot(x, kde_npe(x), color=DEFAULT_COLORS["npe"])
    ax[3].plot(
        x, kde_tabpfn_infomax(x), color=DEFAULT_COLORS["filtered_tabpfn_infomax"]
    )

    for i, a in enumerate(ax):
        a.spines["left"].set_visible(False)
        a.set_yticks([])
        a.set_title(r"$\theta_{}$".format(i + 1))

    plt.legend(
        ["True", "TabPFN", "NPE", "TabPFN-Infomax"],
        loc="lower center",
        ncol=2,
        bbox_to_anchor=(-0.2, -2.9),
        frameon=False,
    )
    fig.tight_layout()
    fig.savefig("lv_long_posterior_marginal_plot.svg", bbox_inches="tight")
```
